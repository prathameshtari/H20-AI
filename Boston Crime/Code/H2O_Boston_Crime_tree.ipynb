```python
import pandas as pd
import numpy as np
import h2o
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
h2o.init()
```

```python
path= 'C:\\Users\\ptari\\Desktop\\INFO 7245\\Assignment_2\\crime_new.csv'
```

```python
hf_df = h2o.import_file(path=path)
hf_df
```

```python
hf_df.isna()
```

```python
hf_df.describe()
```

```python
hf_df=hf_df[['DISTRICT','REPORTING_AREA','MONTH','DAY_OF_WEEK','HOUR','Lat','Long','Day','Night','OFFENSE_CODE_GROUP']]
hf_df.tail()
```

```python
train, valid, test = hf_df.split_frame([0.6, 0.2], seed=1234)

#Prepare predictors and response columns
hf_X = hf_df.col_names[:-1]     #last column is Cover_Type, our desired response variable 
hf_y = hf_df.col_names[-1]    
```

```python
rf_v1 = H2ORandomForestEstimator(
    model_id="rf_covType_v1",
    ntrees=200,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
```

```python
rf_v1.train(hf_X, hf_y, training_frame=train, validation_frame=valid)
```

```python
rf_v1
```

```python
rf_v1.score_history()
```

```python
rf_v1.hit_ratio_table(valid=True)
```

```python
rf_v1.varimp()
```

----------------------------------------------------

```python
# The default depth of trees is 20. It is common to increase this number, to the point that in some implementations, the depth is unlimited. We will increase ours from 20 to 30.  
# 
```

```python
rf_v2 = H2ORandomForestEstimator(
    model_id="rf_covType_v2",
    ntrees=200,
    max_depth=30,
    stopping_rounds=2,
    stopping_tolerance=0.01,
    score_each_iteration=True,
    seed=3000000)
rf_v2.train(hf_X, hf_y, training_frame=train, validation_frame=valid)
```

```python
rf_v2
```

```python
rf_v2.hit_ratio_table(valid=True)
```

```python
final_rf_predictions = rf_v2.predict(test[:-1])
```

```python
final_rf_predictions
```

```python
(final_rf_predictions['predict']==test['OFFENSE_CODE_GROUP']).as_data_frame(use_pandas=True).mean()
```

```python
rf_v2.varimp()
```

```python
gbm_v1 = H2OGradientBoostingEstimator(
    model_id="gbm_covType_v1",
    seed=2000000
)
gbm_v1.train(hf_X, hf_y, training_frame=train, validation_frame=valid)
```

```python
gbm_v1
```

```python
gbm_v1.score_history()
```

```python
gbm_v1.hit_ratio_table(valid=True)
```

```python
gbm_v1_predict=gbm_v1.predict(test[:-1])
```

```python
gbm_v1_predict
```

```python
gbm_v1.hit_ratio_table()
```

```python
(gbm_v1_predict['predict']==test['OFFENSE_CODE_GROUP']).as_data_frame(use_pandas=True).mean()
```

```python
# 1. decrease the number of trees to speed up runtime(from default 50 to 20)
# 2. increase the learning rate (from default 0.1 to 0.2)
# 3. increase the depth (from default 5 to 10)
```

```python
gbm_v2 = H2OGradientBoostingEstimator(
    ntrees=20,
    learn_rate=0.2,
    max_depth=10,
    stopping_tolerance=0.01, #10-fold increase in threshold as defined in rf_v1
    stopping_rounds=2,
    score_each_iteration=True,
    model_id="gbm_covType_v2",
    seed=2000000
)
gbm_v2.train(hf_X, hf_y, training_frame=train, validation_frame=valid)
```

```python
gbm_v2
```

```python
gbm_v2.scoring_history()
```

```python
gbm_v2.hit_ratio_table()
```

```python
# 1. Add a few trees(from 20 to 30)
# 2. Increase learning rate (to 0.3)
# 3. Use a random 70% of rows to fit each tree
# 4. Use a random 70% of columns to fit each tree
```

```python
gbm_v3 = H2OGradientBoostingEstimator(
    ntrees=30,
    learn_rate=0.3,
    max_depth=10,
    sample_rate=0.7,
    col_sample_rate=0.7,
    stopping_rounds=2,
    stopping_tolerance=0.01, #10-fold increase in threshold as defined in rf_v1
    score_each_iteration=True,
    model_id="gbm_covType_v3",
    seed=2000000
)
gbm_v3.train(hf_X, hf_y, training_frame=train, validation_frame=valid)
```
